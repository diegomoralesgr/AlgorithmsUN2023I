# HeapSort Version

```python
# This function returns the index of the parent node
def get_parent_index(node_index):
    return int(node_index / 2)

# This function returns the index of the left child node
def get_left_child_index(node_index):
    return 2 * node_index

# This function returns the index of the right child node
def get_right_child_index(node_index):
    return 2 * node_index + 1

# This function maintains the heap property i.e., parent node is always greater than the child nodes
def maintain_heap_property(heap, index):
  left_child_index = get_left_child_index(index)
  right_child_index = get_right_child_index(index)
  # Compare the node and its left child and find the largest
  if left_child_index <= (len(heap) - 1) and heap[left_child_index] > heap[index]:
      largest_index = left_child_index
  else:
      largest_index = index
  # Compare the right child with the largest value obtained from the previous step
  if right_child_index <= len(heap) - 1 and heap[right_child_index] > heap[largest_index]:
     largest_index = right_child_index
  # Swap the largest value found with the value at the current node
  # If the largest value is not the value at the current node, then heapify the affected subtree
  if largest_index != index:
     heap[index], heap[largest_index] = heap[largest_index], heap[index]
     maintain_heap_property(heap, largest_index)

# This function builds a max heap from an unsorted array
def build_max_heap(unsorted_array):
  # Call max_heapify in reverse level order so that each parent is popped down only once
  for index in range(int(len(unsorted_array) / 2), 0, -1):
      maintain_heap_property(unsorted_array, index)

# Heap Sort
def heap_sort(unsorted_array):
   sorted_array = []
   # Create a max heap from the unsorted array
   build_max_heap(unsorted_array)
   # Move the root of max heap to end of array, reduce heap size and heapify root element
   while len(unsorted_array) > 1:
      max_element = unsorted_array[1]
      unsorted_array[1] = unsorted_array[-1]
      del unsorted_array[-1]
      sorted_array.insert(0, max_element)
      maintain_heap_property(unsorted_array, 1)
   return sorted_array
```

# Heap sort  time complexity histograms for permutations of $n$ different elements counting steps like CLRS I.2.1

```python

def max_heapify_steps(A, i):
    steps = 0
    l = left(i)
    r = right(i)
    if l <= (len(A)- 1) and A[l] > A[i]:
        largest = l
    else:
        largest = i
    if r <= len(A)- 1 and A[r] > A[largest]:
        largest = r
        steps = steps + 1
    if largest != i:
        A[i], A[largest] = A[largest], A[i]
        steps = steps + 1
        steps = steps + max_heapify_steps(A, largest)
    steps = steps + 6
    return steps

def heap_sort_steps(A, B):
    build_max_heap(A)
    steps = 0
    while len(A)>1:
        maxim = A[1]
        A[1] = A[-1]
        del A[-1]
        B.insert(0,maxim)
        steps = steps + 5
        steps = steps + max_heapify_steps(A,1)
    steps = steps + 1
    return steps

```

Generate all permutations without repetitions of size *n* (for small *n*), plot the "exact" time  distribution (number of steps), and print the media and the standar deviation

```python
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np
import math

def heap_sort_steps_by_permutation(n):
  X = range(n)
  times = []

  for P in permutations(X):
    t = heap_sort_steps(list(P), B = [])
    times.append(t)
  b = np.array(range(min(times)-1, max(times)+1)) +0.5
  plt.hist(times, bins = b )
  plt.title("Heap Sort")
  plt.xlabel("Steps")
  plt.ylabel("Frequency")
  plt.show()

  print("n",n)
  print("number of permutations n! =",math.factorial(n))
  print("exp min =",min(times))
  print("exp max =",max(times))
  print("exp mean =",np.mean(times))
  print("stdev",np.std(times))
```

# Average case - expected number of comparisons + swaps on random uniform permutation of $n$   elements with repetitions $t_a(n)$

```python
# Average case on random uniform permtations with repetitions of  $n$   elements  $t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n -\frac{13}{4}= \Theta(n^2) $
import numpy as np
import pandas as pd
import scipy.linalg as la

n = np.array([3.0,4.0,5.0,6.0])    # Number n of sorted elments
t = np.array([23.0, 35.96875, 53.672, 72.45833]) # Average comparisons + swaps for the corresponding n

data = pd.DataFrame({'n': n, 't': t})
data
```

```python
X = np.column_stack([n**0,n**1,n**2,n**3])
a = la.solve(X,t)
a
```
